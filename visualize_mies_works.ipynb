## Visualize the count of the titles of Mies van der Rohe's works in each year:

```python
import pandas as pd
import plotly.express as px

df= pd.read_csv("https://raw.githubusercontent.com/Goli-SF/uzh_demo/main/data/mies.csv")
fig = px.bar(
    df.dropna(subset=["Date"])
      .assign(Date=lambda x: x["Date"].astype(int))
      .groupby("Date", as_index=False)
      .size()
      .sort_values("Date"),
    x="Date",
    y="size",
    labels={"Date": "Year", "size": "Count"},
    title="Count of Records per Year"
)

fig.update_xaxes(tickmode="linear", dtick=1, tickangle=90)
fig.show()
```

## Visualize the count of Mies van der Rohe's unique projects in each year:

```python
clustered_df = pd.read_csv(    "https://raw.githubusercontent.com/Goli-SF/uzh_demo/main/data/mies_clustered.csv")

df_unique = (
    clustered_df
        .dropna(subset=["Date"])
        .assign(Date=lambda x: x["Date"].astype(int))
        .drop_duplicates(subset=["Project", "Date"])
        .groupby("Date", as_index=False)
        .size()
        .sort_values("Date")
)

fig = px.bar(
    df_unique,
    x="Date",
    y="size",
    labels={"Date": "Year", "size": "Unique projects"},
    title="Count of Mies's Unique Projects per Year"
)

fig.update_xaxes(tickmode="linear", dtick=1, tickangle=90)
fig.show()
```
