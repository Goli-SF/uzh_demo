```python
import pandas as pd
import plotly.express as px

df= pd.read_csv("./data/mies.csv")
fig = px.bar(
    df.dropna(subset=["Date"])
      .assign(Date=lambda x: x["Date"].astype(int))
      .groupby("Date", as_index=False)
      .size()
      .sort_values("Date"),
    x="Date",
    y="size",
    labels={"Date": "Year", "size": "Count"},
    title="Count of Records per Year"
)

fig.update_xaxes(tickmode="linear", dtick=1, tickangle=90)
fig.show()
```
